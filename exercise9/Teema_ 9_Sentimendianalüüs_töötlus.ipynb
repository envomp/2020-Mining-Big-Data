```python
# Testi andmeraamistiku järjekorra segamist

import numpy as np
import pandas as pd
import os

df = pd.DataFrame([("jkaj ukad kkkk", 1),
                   ("jadakk", 1),
                   ("adaf jadakk", 0),
                   ("adadakk", 0),
                  ])
print(df)
print(df.index)
np.random.seed(0)
#print(np.random.permutation(df.index))
#df = df.reindex(np.random.permutation(df.index)).sort_index()
df = df.sample(frac=1.0)
print(df)
```

```python


# Eeldus: http://ai.stanford.edu/~amaas/data/sentiment/ konteineri sisu
# on lahti pakitud selle märkmikuga samasse kataloogi

count = 0 # loetud failide loendur, progressi näitamiseks

# Loe failid andmeraamistikku
labels = {'pos': 1, 'neg': 0}
df = pd.DataFrame()

for s in ('test', 'train'):
    for l in labels:
        path = './aclImdb/%s/%s' % (s, l)
        for fn in os.listdir(path):
            with open(os.path.join(path, fn), 'r', encoding="utf-8") as infile:
                txt = infile.read()
            df = df.append([[txt, labels[l]]], ignore_index=True)
            
            count += 1
            if count % 1000 == 0: print(count)
            
df.columns = ['review', 'sentiment']

print(df.head(3))

# Sega juhuslikult andmeraamistiku read
#np.random.seed(1)
#df = df.reindex(np.random.permutation(df.index)).sort_index()
df = df.sample(frac=1.0)

print("Klassid:", df.sentiment.head(10))

# Kirjuta andmeraamistik csv faili
df.to_csv('./movie_data.csv', index=False)
```
